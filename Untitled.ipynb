```python
import os
import subprocess
import matplotlib.pyplot as plt
import time
 
```

```python
bashCommand = "tail -n 1 /dev/ttyACM0"
values={"presion":[] ,"hum":[] ,"temp1":[], "temp2":[]}
```

```python
for i in range(20):
    values["presion"].append(0)
    values["hum"].append(0)
    values["temp1"].append(0)
    values["temp2"].append(0)
    
```

```python
%matplotlib inline
import time
import pylab as pl
from IPython import display
for i in range(10):
    pl.plot(pl.randn(100))
    display.clear_output(wait=True)
    display.display(pl.gcf())
    pl.clf()

    time.sleep(1.0)
```

```python
%matplotlib inline
plt.ion()

while True:
    val= subprocess.check_output(bashCommand, shell=True).split(",")
    if(len(val)==4):
        values["presion"].append(val[0])
        values["hum"].append(val[1])
        values["temp1"].append(val[2])
        values["temp2"].append(val[3])
        time.sleep(7)

        plotValues(values)    
```

```python
def plotValues(values):
#     pl.plot(pl.randn(100))
#     display.clear_output(wait=True)
#     display.display(pl.gcf())
#     pl.clf()

#     time.sleep(1.0)
    
    
    
    
    x=range(20)
    plt.clf()
    f, ((ax1, ax2), (ax3, ax4)) = pl.subplots(2, 2)
    ax1.plot(x, values["presion"][-21:-1])
    #ax2.scatter(x, values["hum"][-21:-1])
    ax3.scatter(x, values["temp1"][-21:-1], color='r')
    ax4.plot(x, values["temp2"][-21:-1], color='r')
    display.clear_output(wait=True)
    display.display(pl.gcf())
    pl.clf()

    time.sleep(1.0)
```
